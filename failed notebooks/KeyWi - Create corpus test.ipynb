# Import titles to a new column

```python
from bs4 import BeautifulSoup
import re
```

```python
# def cleanText(text):
#   text = text.replace('|', '')
#   text = text.replace('-', '')
#   text = text.replace('\n', ' ')
#   text = text.strip()
#   return text

def cleanText(text):
    # text = BeautifulSoup(text, "lxml").text
    text = re.sub(r'\|\|\|', r' ', text) 
    text = re.sub(r'http\S+', r'<URL>', text)
    text = text.lower()
    text = text.replace('x', '')
    return text

text_list = []
for filename in tqdm(balanced_dataset.raw_filename):
  with open(filename) as fp:
    soup = BeautifulSoup(fp, 'html.parser')
  fp.close()
  # for text in soup.find_all('p'):
  # if soup.title is not None else 'unknown'
  for paragraph in soup.find_all('p'):
    text_list.append(cleanText(paragraph.get_text()))
  break
    


  # break

text_list
```

```python


clean_text = []
for text in text_list:
  clean_text.append(cleanText(text))
```

```python
clean_text
```

```python
balanced_dataset['title'] = clean_text
```
